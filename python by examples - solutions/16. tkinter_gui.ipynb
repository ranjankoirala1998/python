<h3>Challenge 126</h3>
<p>Create a program that will ask the user to enter a number in a box. When they click on a button it will add that number to a total and display it in another box. This can be repeated as many times as they want and keep adding to the total. There should be another button that resets the total back to 0 and empties the original text box, ready for them to start again.</p>

```python
from tkinter import *

def add() :
    global total
    
    try :
        num = int(entry.get())
        total += num
    except :
        total += 0

    entry.delete(0, END)
    
    msg.pack(side= BOTTOM)
    msg["text"] = "Total : " + str(total)

def clear() :
    global total
    total = 0
    entry.delete(0, END)
    msg.pack_forget()

    

window = Tk()
window.title("#126")
window.geometry("300x300")

total = 0

ask = Label(window, text="Enter a number : ")
ask.place(x=25, y=25)

entry = Entry(window, text=0)
entry.place(x=125, y=25, width=100, height=25)
entry["justify"] = "center"
entry.focus()

add_btn = Button(window, text='ADD', command=add)
add_btn.place(x=125, y=50, width=50, height=25)

clear_btn = Button(window, text='CLEAR', command=clear)
clear_btn.place(x=175, y=50, width=50, height=25)

msg = Message(window, text = "", width= 100)
msg.place(x=125, y=100, width=100, height=50)
msg.pack(side=BOTTOM)

window.mainloop()
```

<h3>Challenge 130</h3>
<p>Alter program 129 to add a third button that will save the list to a .csv file. The code  tmp_list = num_list.get(0,END) can be used to save the contents of a list box as a tuple called tmp_list.</p>

```python
from tkinter import *
import csv

def is_whole() :
    num = entry.get()
    if num.isdigit() : 
        num_list.insert(END, num)
    entry.delete(0, END)
    entry.focus()
        

def clear() :
    num_list.delete(0, END)
    entry.delete(0, END)
    entry.focus()

def save_as_csv() :
    tmp_list = num_list.get(0,END)

    try :
        whole_numbers = open("files/WholeNumbers.csv", "r")
        whole_numbers.close()

    except : 
        whole_numbers = open("files/WholeNumbers.csv", "w")
        whole_numbers.close()
    
    whole_numbers = open("files/WholeNumbers.csv", "a")
    
    row = ""
    for num in tmp_list : 
        if tmp_list.index(num) == len(tmp_list)-1 :
            row = row + num + '\n'
        else : 
            row = row + num + ','
    
    whole_numbers.write(row)
    whole_numbers.close()


window = Tk()
window.title("#130")
window.geometry("450x250")

label = Label(window, text= "Enter a number : ")
label.place(x= 25, y= 25, width= 100, height= 25)

entry = Entry(window)
entry.place(x= 125, y= 25, width= 100, height= 25)
entry["justify"] = "center"
entry.focus()

num_list = Listbox(window)
num_list.place(x= 125, y=50, width= 100, height= 150)

is_whole = Button(window, text= "Check and Add", command= is_whole)
is_whole.place(x= 230, y= 25, width= 100, height= 25)

clear = Button(window, text= "Clear", command= clear)
clear.place(x= 335, y= 25, width= 50, height= 25)

save = Button(window, text= "Save As CSV", command= save_as_csv)
save.place(x= 125, y= 200, width= 100, height= 25)

window.mainloop()
```

<h3>Challenge 131</h3>
<p>Create a program that will allow the user to create a new .csv file. It should ask them to enter the name and age of a person and then allow them to add this to the end of the file they have just created.</p>

```python
from tkinter import *
import csv

def add_to_csv() : 
    try :
        name_age = open("files/NameAge.csv", "r")
        name_age.close()

    except : 
        name_age = open("files/NameAge.csv", "w")
        name_age.close()
    
    name_age = open("files/NameAge.csv", "a")

    name = entry1.get()
    age = entry2.get()

    if age.isdigit() :
        row = name + ',' + age + '\n'
        name_age.write(row)
    
    entry1.delete(0, END)
    entry1.focus()
    entry2.delete(0, END)
    entry1.focus()

    name_age.close()


window = Tk()
window.title("#131")
window.geometry("500x250")

label1 = Label(window, text= "Enter a name : ", width= 100)
label1.place(x= 25, y= 25, width= 100, height=25)

entry1 = Entry(window)
entry1.place(x= 125, y= 25, width= 125, height=25)
entry1["justify"] = "center"
entry1.focus()

label2 = Label(window, text= "Enter an age : ", width= 100)
label2.place(x= 25, y= 50, width= 100, height=25)

entry2 = Entry(window)
entry2.place(x= 125, y= 50, width= 125, height=25)
entry2["justify"] = "center"
entry2.focus()

save = Button(window, text= "Add To CSV", command= add_to_csv)
save.place(x= 125, y= 75, width= 125, height=25)

window.mainloop()
```

```python
from tkinter import *
import csv

def main() :

    def add_to_csv() : 
        try :
            name_age = open("files/NameAge.csv", "r")
            name_age.close()

        except : 
            name_age = open("files/NameAge.csv", "w")
            name_age.close()
        
        name_age = open("files/NameAge.csv", "a")

        name = entry1.get()
        age = entry2.get()

        if age.isdigit() :
            row = name + ',' + age + '\n'
            name_age.write(row)
        
        entry1.delete(0, END)
        entry1.focus()
        entry2.delete(0, END)
        entry1.focus()

        name_age.close()

    def import_from_csv() :
        frame.destroy()

        frame2 = Frame(window, width=400, height= 200)
        frame2.place(x=0, y=0)

        list_box = Listbox(frame2)
        list_box.place(width= 200, height= 150)


        try :
            name_age = open("files/NameAge.csv", "r")
            name_age.close()

        except : 
            name_age = open("files/NameAge.csv", "w")
            name_age.close()
        
        name_age = open("files/NameAge.csv", "r")
        records = list(csv.reader(name_age))
        name_age.close()
        for record in records :
            row = record[0] + ',' + record[1]
            list_box.insert(END, row)

        main_menu = Button(frame2, text= "Main Window", command= main)
        main_menu.place(x= 50, y= 150, width= 100, height=25)
    
    frame = Frame(window, width=400, height= 200)
    frame.place(x=0, y=0)

    label1 = Label(frame, text= "Enter a name : ", width= 100)
    label1.place(x= 25, y= 25, width= 100, height=25)

    entry1 = Entry(frame)
    entry1.place(x= 125, y= 25, width= 180, height=25)
    entry1["justify"] = "center"
    entry1.focus()

    label2 = Label(frame, text= "Enter an age : ", width= 100)
    label2.place(x= 25, y= 50, width= 100, height=25)

    entry2 = Entry(frame)
    entry2.place(x= 125, y= 50, width= 180, height=25)
    entry2["justify"] = "center"
    entry2.focus()

    save = Button(frame, text= "Add To CSV", command= add_to_csv)
    save.place(x= 165, y= 75, width= 100, height=25)

    import_from_csv = Button(frame, text= "Import From CSV", command= import_from_csv)
    import_from_csv.place(x= 165, y= 150, width= 100, height=25)

    frame.pack()


window = Tk()
window.title("#131")
window.geometry("400x200")

main()

window.mainloop()
```
