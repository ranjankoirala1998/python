<h3>The Challenge</h3>
<p>A shift code is where a message can be easily encoded and is one of the simplest codes to use. Each letter is moved forwards through the alphabet a set number of letters to be represented by a new letter. For instance, “abc” becomes “bcd” when the code is shifted by one (i.e. each letter in the alphabet is moved forward one character).</p>
<p>You need to create a program which will display the following menu: </br></br>1) Make a code</br>2) Decode a message</br>3) Quit</br></br>Enter your selection : </p>
<p>If the user selects 1, they should be able to type in a message (including spaces) and then enter a number. Python should then display the encoded message once the shift code has been applied.</br></br> 
If the user selects 2, they should enter an encoded message and the correct number and it should display the decoded message (i.e. move the correct number of letters backwards through the alphabet).</br></br>
If they select 3 it should stop the program from running.</br></br>
After they have encoded or decoded a message the menu should be displayed to them again until they select quit. </p>
<h3>Problem You Will Have To Overcome</h3>
<p>Decide if you want to allow both upper and lower case letters or if you want to convert all the data into one case.</br></br>
Decide if you are allowing punctuation.</br></br>
If the shift makes the letter go past the end of the alphabet it should start again; i.e. if the user enters “xyz” and 5 is entered as the shift number, it should display “bcd”. This should work the opposite way for decoding a message, so if the value gets to “a” it will go back to “w”.</br></br>
Make sure that suitable messages are displayed if the user selects an inappropriate option on the menu or selects an inappropriate number to make the shift code.</br></br>
Test out your decode option by decoding the message “we ovugjohsslunl”, which was created with the number 7 when the code only uses “abcdefghijklmnopqrstuvwxyz ” (note the space at the end).</p>

```python
lower_letters = 'abcdefghijklmnopqrstuvwxyz'
upper_letters = lower_letters.upper()

def encode() :

    msg = input("Enter the message to be encoded : ")
    shift = int(input("Enter a number of shifts : "))
    encoded = ""
    for letter in msg :
        if letter in lower_letters :
            i = lower_letters.index(letter)
            j = (i + (shift % 26)) % 26
            encoded += lower_letters[j]

        elif letter in upper_letters :
            i = upper_letters.index(letter)
            j = (i + (shift % 26)) % 26
            encoded += upper_letters[j]
        
        else :
            encoded += letter
    
    print("Encoded Message : {}".format(encoded))


def decode() :
    
    msg = input("Enter the message to be decoded : ")
    shift = int(input("Enter a number of shifts : "))
    decoded = ""
    for letter in msg :
        if letter in lower_letters :
            i = lower_letters.index(letter)
            j = (i - (shift % 26)) % 26
            decoded += lower_letters[j]

        elif letter in upper_letters :
            i = upper_letters.index(letter)
            j = (i - (shift % 26)) % 26
            decoded += upper_letters[j]
        
        else :
            decoded += letter
    
    print("Decoded Message : {}".format(decoded))


while True :

    print("""
    ------- MENU -------
    1) Make a code
    2) Decode a message
    3) Quit
    """)

    choice = int(input("Enter your selection : "))

    if choice == 1 :
        encode()

    elif choice == 2 :
        decode()

    elif choice == 3 :
        quit()

    else : 
        print("INVALID OPTION")
```
