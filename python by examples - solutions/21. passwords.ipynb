<h3>The Challenge</h3>
<p>You need to create a program that will store the user ID and passwords for the users of a system. It should display the following menu: </br></br>1) Create a new User ID</br>2) Change a password</br>3) Display all User IDs</br>4) Quit</br></br>Enter selection : </p>
<p>If the user selects 1, it should ask them to enter a user ID. It should check if the user ID is already in the list. If it is, the program should display a suitable message and ask them to select another user ID. Once a suitable user ID has been entered it should ask for a password. Passwords should be scored with 1 point for each of the following:</br></br>  
 it should have at least 8 characters;</br>   it should include uppercase letters;</br>   it should include lower case letters;</br>   it should include numbers; and</br>   it should include at least one special character such as !, £, $, %, &, <, * or @.</br></br>If the password scores only 1 or 2 it should be rejected with a message saying it is a weak password; if it scores 3 or 4 tell them that “This password could be improved.” Ask them if they want to try again. If it scores 5 tell them they have selected a strong password. Only acceptable user IDs and passwords should be added to the end of the .csv file.</br></br>
If they select 2 from the menu they will need to enter a user ID, check to see if the user ID exists in the list, and if it does, allow the user to change the password and save the changes to the .csv file. Make sure the program only alters the existing password and does not create a new record.</br></br>
If the user selects 3 from the menu, display all the user IDs but not the passwords.</br></br>
If the user selects 4 from the menu it should stop the program.</p>
<h3>Problems You Will Have To Overcome</h3>
<p>As existing data in .csv files cannot be edited and can only be read or added to, you will need to import the data as a temporary list into Python so you can make the changes before the data is written to the .csv file afresh.</br></br> 
Make sure only passwords belonging to an existing user ID can be altered.</br></br> 
Use suitable messages to guide the user easily through the system.</br></br>  
Repeat the menu until they quit the program.</p>

```python
import csv

def get_user() :
    try :
        passwords = open("files/Passwords.csv", "r")
        passwords.close()

    except : 
        passwords = open("files/Passwords.csv", "w")
        passwords.write('userID' + ',' + 'password' + '\n')
        passwords.close()


    passwords = open("files/Passwords.csv", "r")
    records = list(csv.reader(passwords))
    passwords.close()

    user = input("Enter User ID : ") 

    flag = True 

    while flag :

        for record in records :
            if user in record :
                flag = True
                break
            else :
                flag = False
                continue 
        
        if flag : 
            print("THIS USER ID ALREADY EXISTS")
            user = input("Enter a new User ID : ")         
                
                
    
    pwd = input("Enter a password : ")
    new_pwd = check_pwd(pwd)

    passwords = open("files/Passwords.csv", "a")
    passwords.write(user +','+ new_pwd +'\n')
    passwords.close()
    

def change_pwd() :
    passwords = open("files/Passwords.csv", "r")
    records = list(csv.reader(passwords))
    passwords.close()

    user = input("Enter User ID : ") 

    flag = True 

    while flag :

        for record in records :
            if user in record :
                index = records.index(record)
                flag = False
                break
            else :
                flag = True
                continue 

        if flag : 
            print("THIS USER ID DOESN'T EXIST")
            user = input("Enter another User ID : ") 

        else : 
            password = input("Enter new password :")
            records[index][1] = check_pwd(password)
            print("YOUR PASSWORD HAS BEEN CHANGED") 


    passwords = open("files/Passwords.csv", "w")

    for record in records :
        passwords.write(record[0] + ',' + record[1] + '\n')
    
    passwords.close()


def display_users() :
    passwords = open("files/Passwords.csv", "r")
    records = list(csv.reader(passwords))
    passwords.close()

    for record in records : 
        print(record[0] + ' , ' + record[1])


def check_pwd(password) :
    
    pwd = password

    score = 0

    if len(pwd) > 7 :
        score += 1

    for char in pwd :
        if char.islower() :
            score += 1
            break

    for char in pwd :
        if char.isupper() :
            score += 1
            break

    for char in pwd :
        if char.isdigit() :
            score += 1
            break

    for char in pwd :
        if char in '!£$%&<*@' :
            score += 1
            break

    if score < 3 : 
        print("PASSWORD TOO WEAK")
        pwd = input("Enter a new password : ")
        pwd = check_pwd(pwd)
    
    elif score == 3 or score == 4 :
        print("THIS PASSWORD COULD BE IMPROVED")
        choice = input("Do you want to improve your password ? y/n : ")
        if choice == 'y' :
            pwd = input("Enter a new password : ")
            pwd = check_pwd(pwd) 
    else : 
        print("YOU HAVE SELECTED A STRONG PASSWORD")

    return pwd
    


flag = True

while flag :
    print("""
----------MENU----------
1) Create a new User ID
2) Change a password
3) Display all User IDs
4) Quit
    """)
    selection = int(input("Enter selection : "))

    if selection == 1 :
        get_user()

    elif selection == 2 :
        change_pwd()

    elif selection == 3 :
        display_users()

    elif selection == 4 :
        flag = False

    else : 
        print("YOUR SELECTION WAS INVALID. TRY AGAIN !!")
```
