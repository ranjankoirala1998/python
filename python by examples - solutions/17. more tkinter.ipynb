<h3>Challenge 135</h3>
<p>Create a simple program that shows a drop-down list containing several colours and a Click Me button. When the user selects a colour from the list and clicks the button it should change the background of the window to that colour. For an extra challenge, try to avoid using an if statement to do this.</p>

```python
from tkinter import *

def click_me() :
    color = selectColor.get()
    if color == "Select Color" :
        pass

    else :
        window.configure(background= color)

window = Tk()
window.title("#135")
window.geometry("400x400")

selectColor = StringVar(window)
selectColor.set("Select Color")
colorList = OptionMenu(window, selectColor, "red", "blue", "green", "yellow", "purple")
colorList.place(x = 150, y = 150, width= 100, height= 25) 

clickMe = Button(window, text= "CLICK ME", command= click_me)
clickMe.place(x = 162.5, y = 175, width= 75, height= 25)


window.mainloop()
```

<h3>Challenge 137</h3>
<p>Change program 136 so that when a new name and gender is added to the list box it is also written to a text file. Add another button that will display the entire text file in the main Python shell window.</p>

```python
from tkinter import *

def add_me() :

    name = name_entry.get()
    gender = selectGender.get()
    row = name + ' , ' + gender

    nameList.insert(END, row)
    name_entry.delete(0, END)

    file = open('files/add_me.txt', 'a')
    file.write(row + '\n')
    file.close()

def display() :
    file = open('files/add_me.txt', 'r')
    print(file.read())
    file.close()

window = Tk()
window.title("#135")
window.geometry("400x400")

file = open('files/add_me.txt', 'w')
file.close()

label_name = Label(window, text= "Enter your name : ", width= 100)
label_name.place(x=25, y= 25, width= 125, height= 25)

name_entry = Entry(window)
name_entry.place(x=150, y= 25, width= 125, height= 25)
name_entry["justify"] = "center"
name_entry.focus()

label_gender = Label(window, text= "Select your gender : ", width= 125)
label_gender.place(x=25, y= 50, width= 125, height= 25)

selectGender = StringVar(window)
selectGender.set("Male")
genderList = OptionMenu(window, selectGender, "Male", "Female")
genderList.place(x = 150, y = 50, width= 125, height= 25) 

addMe = Button(window, text= "ADD ME", command= add_me)
addMe.place(x = 150, y = 75, width= 75, height= 25)

label_namelist = Label(window, text= "NAME LIST", width= 100)
label_namelist.place(x=25, y= 125, width= 250, height= 25)

nameList = Listbox(window)
nameList.place(x = 25, y = 150, width= 250, height= 150)

displayList = Button(window, text= "Display List From File", command= display)
displayList.place(x = 87.5, y = 325, width= 125, height= 25)

window.mainloop()
```
