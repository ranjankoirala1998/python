<p>Define a subprogram that will ask the user to enter a number and save it as the variable “num”. Define another subprogram that will use “num” and count from 1 to that number.</p>

```python
def ask_num() :
    num = int(input("Enter a number : "))
    
    return num

def counter(num) :
    for i in range(num) : 
        print(i+1)

number = ask_num()
counter(number)
```

<h3>Challenge 119</h3>
<p>Define a subprogram that will ask the user to pick a low and a high number, and then generate a random number between those two values and store it in a variable called “comp_num”.</br></br>
 
Define another subprogram that will give the instruction “I am thinking of a number…” and then ask the user to guess the number they are thinking of.</br></br>
 
Define a third subprogram that will check to see if the comp_num is the same as the user’s guess. If it is, it should display the message “Correct, you win”, otherwise it should keep looping, telling the user if they are too low or too high and asking them to guess again until they guess correctly. </p>

```python
import random

def random_generator() :
    lower = int(input("Enter the lower limit : "))
    upper = int(input("Enter the upper limit : "))
    
    comp_num = random.randint(lower, upper)
    
    return comp_num

def number_thinker() :
    guess = int(input("I am thinking of a number ..."))
    
    return guess

def check_guess(guess, number) : 
    flag = True
    while flag :
        if guess == number :
            print("Correct!")
            flag = False
    
        else : 
            if guess > number :
                print("Too High, Try Again!") 

            else :
                print("Too Low, Try Again!") 

            guess = number_thinker()

number = random_generator()
guess = number_thinker()
check_guess(guess, number)
```

<h3>Challenge 120</h3>
<p>Display the following menu to the user:</br>1) Addition</br>2) Subtraction</br>Enter 1 or 2 : </p>

<p>If they enter a 1, it should run a subprogram that will generate two random numbers between 5 and 20, and ask the user to add them together. Work out the correct answer and return both the user’s answer and the correct answer.</br></br>
 
If they entered 2 as their selection on the menu, it should run a subprogram that will generate one number between 25 and 50 and another number between 1 and 25 and ask them to work out num1 minus num2. This way they will not have to worry about negative answers. Return both the user’s answer and the correct answer.</br></br>
 
Create another subprogram that will check if the user’s answer matches the actual answer. If it does, display “Correct”, otherwise display a message that will say “Incorrect, the answer is” and display the real answer.</br></br> 
 
If they do not select a relevant option on the first menu you should display a suitable message.</p>

```python
import random

def add_random() :
    num1 = random.randint(5, 20)
    num2 = random.randint(5, 20)
    answer = num1 + num2
    guess = int(input("{} + {} = ".format(num1, num2)))
    check_answer(answer, guess)


def sub_random() :
    num1 = random.randint(25, 50)
    num2 = random.randint(1, 25)
    answer = num1 - num2
    guess = int(input("{} - {} = ".format(num1, num2)))
    check_answer(answer, guess)


def check_answer(answer, guess) :
    if answer == guess :
        print("Correct!")

    else :
        print("Incorrect, the answer is {}.".format(answer))


print("""
1) Addition
2) Subtraction\n
""")

choice = int(input("Enter 1 or 2 : "))

if choice == 1 :
    add_random()

elif choice == 2 :
    sub_random()

else :
    print("Invalid Option!!")
```

<h3>Challenge 121</h3>
<p>Create a program that will allow the user to easily manage a list of names. You should display a menu that will allow them to add a name to the list, change a name in the list, delete a name from the list or view all the names in the list. There should also be a menu option to allow the user to end the program. If they select an option that is not relevant, then it should display a suitable message. After they have made a selection to either add a name, change a name, delete a name or view all the names, they should see the menu again without having to restart the program. The program should be made as easy to use as possible.</p>

```python
def add_name(names) :
    name = input("Enter a name : ")
    if name not in names :
        names.append(name)

    else : 
        print("The name '{}' already exists in the list.".format(name))


def change_name(names) :
    name  = input("Which name do you want to change? Enter name : ")
    if name in names :
         idx = names.index(name)
         new_name = input("Enter new name : ")
         names[idx] = new_name


    else : 
        print("The name '{}' does not exist in the list.".format(name))
    

def delete_name(names) :
    name  = input("Which name do you want to remove? Enter name : ")
    if name in names :
        names.remove(name)

    else : 
        print("The name '{}' does not exist in the list.".format(name))


def view_all(names) :
    if len(names) == 0 :
        print("List is empty.")

    else : 
        for name in names : 
            print(name)


def end_program():
    print("SEE YOU SOON.")
    return False


namelist = []

flag = True

while flag : 

    print('''
    -------------------------------
                OPTIONS
    -------------------------------
    1. Add a name to the list
    2. Change a name in the list
    3. Delete a name from the list
    4. View all names in the list
    5. Quit the program\n
    ''')

    choice = int(input("Enter an option (1-5) : "))

    if choice == 1 : 
        add_name(namelist)

    elif choice == 2 :
        change_name(namelist)

    elif choice == 3 : 
        delete_name(namelist)

    elif choice == 4 : 
        view_all(namelist)

    elif choice == 5 :
        flag = end_program() 

    else : 
        print("Invalid Option!!")
```

<h3>Challenge 122</h3>
<p>Display the following menu to the user:</br>1) Add to file</br>2) View all records</br>3) Quit program</br></br>Enter the number of your selection : </p>

<p>If the user selects 1, allow them to add to a file called Salaries.csv which will store their name and salary. If they select 2 it should display all records in the Salaries.csv file. If they select 3 it should stop the program. If they select an incorrect option they should see an error message. They should keep returning to the menu until they select option 3.</p>

```python
import csv

def add_record() :
    salaries = open("files/Salaries.csv", "a")
    name = input("Enter name : ")
    salary = input("Enter salary : ")
    salaries.write(name+','+salary+'\n')
    salaries.close()


def view_record() :
    try :
        salaries = open("files/Salaries.csv", "r")
        salaries.close()

    except : 
        salaries = open("files/Salaries.csv", "w")
        salaries.close()

    salaries = open("files/Salaries.csv", "r")
    records = list(csv.reader(salaries))

    if len(records) == 0 :
        print("There are no any records.")
    
    else : 
        print("""
        ---------------
            RECORDS
        ---------------""")
        for record in records :
            print(record[0]+','+record[1])
    
    salaries.close()


def end_program():
    print("SEE YOU SOON.")
    return False


flag = True

while flag : 

    print('''
    -------------------------------
                OPTIONS
    -------------------------------
    1) Add to file
    2) View all records
    3) Quit program\n
    ''')

    choice = int(input("Enter an option (1-3) : "))

    if choice == 1 : 
        add_record()

    elif choice == 2 :
        view_record()

    elif choice == 3 :
        flag = end_program() 

    else : 
        print("Invalid Option!!")
```

<h3>Challenge 123</h3>
<p>In Python, it is not technically possible to directly delete a record from a .csv file. Instead you need to save the file to a temporary list in Python, make the changes to the list and then overwrite the original file with the temporary list.</p>
<p>Change the previous program to allow you to do this. Your menu should now look like this : </br></br>1) Add to file</br>2) View all records</br>3) Delete a record</br>4) Quit program</br></br>Enter the number of your selection :</p>

```python
import csv

def add_record() :
    salaries = open("files/Salaries.csv", "a")
    name = input("Enter name : ")
    salary = input("Enter salary : ")
    salaries.write(name+','+salary+'\n')
    salaries.close()


def view_record() :
    try :
        salaries = open("files/Salaries.csv", "r")
        salaries.close()

    except : 
        salaries = open("files/Salaries.csv", "w")
        salaries.close()

    salaries = open("files/Salaries.csv", "r")
    records = list(csv.reader(salaries))

    if len(records) == 0 :
        print("There are no any records.")
    
    else : 
        print("""
        ---------------
            RECORDS
        ---------------""")
        for record in records :
            print(record[0]+','+record[1])
    
    salaries.close()


def delete_record() :
    try :
        salaries = open("files/Salaries.csv", "r")
        salaries.close()

    except : 
        salaries = open("files/Salaries.csv", "w")
        salaries.close()

    salaries = open("files/Salaries.csv", "r")
    records = list(csv.reader(salaries))
    salaries.close()

    if len(records) == 0 :
        print("There are no any records.")
    
    else : 
        row = int(input("Which row do you want to remove? Enter row number : "))
        
        if row <= len(records) :
            del records[row-1]
        
        else : 
            print("No any record in the row.")

    
    salaries = open("files/Salaries.csv", "w")

    for record in records : 
        salaries.write(record[0]+','+record[1]+'\n')

    salaries.close()


def end_program():
    print("SEE YOU SOON.")
    return False


flag = True

while flag : 

    print('''
    -------------------------------
                OPTIONS
    -------------------------------
    1) Add to file
    2) View all records
    3) Delete a record
    4) Quit program\n
    ''')

    choice = int(input("Enter an option (1-3) : "))

    if choice == 1 : 
        add_record()

    elif choice == 2 :
        view_record()

    elif choice == 3 :
        delete_record() 
    
    elif choice == 4 :
        flag = end_program() 

    else : 
        print("Invalid Option!!")
```
